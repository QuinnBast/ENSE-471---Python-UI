In this portion of the workbook, you will learn to use the line command to draw a straight line on the canvas. First we must access the canvas portion of the widget. All widgets, including layouts, have a canvas component. The purpose of the canvas is to allow draw commands. To access this in kivy, we first use the "canvas:" keyword. We may then make draw calls, including lines, rectangles and ellipses.


## Part 5: Draw a line based on touch events

Now that we have a good handle on drawing lines from the console, lets try getting user input to draw lines as a set of continuous points. Recall that a line is defined as a list of different coordinates which alternate between x and y coordinates. If we can get the position that the user is touching on the screen and add this location to the list of points, we can then create a simple touch drawing application.

You can read about touch inputs at the following tutorial:

https://kivy.org/docs/guide/inputs.html

It is worth noting that touch input also applies to mouse clicks.

The key take aways is:
    
    On any widget you can define the following functions:
        - on_touch_down
        - on_touch_move
        - on_touch_up
    These functions trigger automatically when the user touches, moves, and releases the interface. Inside of the function we may access:
    - touch.x - The x coordinate of the touch
    - touch.y - The y coordinate of the touch
    - touch.ud - a dictionary which allows setting custom attributes for a touch. 
        
The following code is based on the tutorial here:

https://kivy.org/docs/tutorials/firstwidget.html

## Exercise:

Complete the following demo. You need to finish the two commented lines of code. In the first line of code you will create a new line object using the current mouse position. In the second line you will complete the line using a second set of coordinates.

````python
from random import random
from kivy.app import App
from kivy.uix.widget import Widget
from kivy.graphics import Color, Line

from kivy.config import Config
Config.set('graphics', 'fullscreen', '0')


class MyPaintWidget(Widget):

    def on_touch_down(self, touch):
        color = (random(), random(), random(), 1)
        with self.canvas:
            Color(*color, mode='rgba')
            line = Line()
            line.points = [touch.x, touch.y]
            line.width = 2
            touch.ud['line'] = line# Create a new line here with points set to touch.x and touch.y, width = 2

    def on_touch_move(self, touch):
        touch.ud['line'].points += [touch.x, touch.y]# append the current cursor position using the touch.x and touch.y variables


class MyPaintApp(App):

    def build(self):
        parent = Widget()
        self.painter = MyPaintWidget()
        parent.add_widget(self.painter)
        return parent

if __name__ == '__main__':
    MyPaintApp().run()
````

Note: In this example we built our app using a custom build command. This will be necessary to prevent the draw commands from affecting buttons that we add to the interface.
